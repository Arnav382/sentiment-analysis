#Event extraction

```python
import spacy
nlp=spacy.load("en")
sent_vecs={}
docs=[]
```

```python
for title in tqdm(df.title):
  doc=nlp(title)
  docs.append(doc)
  sent_vecs.update({title:doc.vector})
  sentences=list(sent_vecs.keys())
  vectors=list(sent_vecs.values())
```

```python
# vectors
```

### clustering those vectors

```python
import numpy as np
x=np.array(vectors)
n_classes={}
```

```python
for i in tqdm(np.arange(0.001,1,0.002)):
  dbscan=DBSCAN(eps=i,min_samples=2,metric="cosine").fit(x)
  n_classes.update({i:len(pd.Series(dbscan.labels_).value_counts())})
  dbscan=DBSCAN(eps=.08,min_samples=2,metric="cosine").fit(x)
```

Converting string time column to date_time format for lining up events

```python
k=0
jst=[]
for i in df["date_time"]:
  mystr=i
  my_date = dt.strptime(mystr, "%Y-%m-%d %H:%M:%S")
  jst.append(my_date)
```

```python
df["timestamp"]=jst
```

```python
lst_clusters=[]
for i in np.unique(dbscan.labels_):
  results=pd.DataFrame()
  results = pd.DataFrame({"label":dbscan.labels_,"sent":sentences})
  example_result=pd.DataFrame()
  example_result=results[results.label==i].sent.tolist()
  event_df=pd.DataFrame()
  event_df=df[df.title.isin(example_result)][["timestamp","title"]]
  event_df=event_df.sort_values(by="timestamp").dropna()
  lst_clusters.append(event_df)

  
```
